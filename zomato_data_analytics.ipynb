Average Rating by Restaurant Type
Group by restaurant type and calculate the mean of rate (out of 5).
Insight: Which restaurant types tend to have higher ratings?


```python
#group by restaurent type and calculate mean of ratings
```

```python
avg_rating_by_type=df.groupby('restaurant type')['rate (out of 5)'].mean().sort_values(ascending=False)
avg_rating_by_type
```

```python
avg_rating_by_type=df.groupby('restaurant type').agg({'rate (out of 5)':'mean'}).sort_values(by='rate (out of 5)',ascending=False)[:20]
avg_rating_by_type
```

```python
#visualization
```

```python
avg_rating_by_type.plot(kind='barh',title='Average Rating by Restaurant Type',ylabel='Restaurant Type',xlabel='Average Rating')
plt.show()
```

```python
#Total Number of Ratings by Area
#Group by area and sum the num of ratings.
#Insight: Which areas have the most customer engagement based on ratings?
```

```python
rating_by_area=df.groupby('area').agg({'num of ratings':'sum'}).sort_values(by='num of ratings',ascending=False)
rating_by_area
```

```python
#visualize
```

```python
rating_by_area.plot(kind='barh',title='Total Number of Ratings by Area',ylabel='Area',xlabel='Total Ratings')
plt.show()
```

```python
#Average Cost by Cuisine Type
#Group by cuisines type and compute the average of avg cost (two people).
#Insight: Which cuisines are more expensive?
```

```python
avg_cost_by_cuisine=df.groupby('cuisines type').agg({'avg cost (two people)':'mean'}).sort_values(by='avg cost (two people)',ascending=False)[:25]
avg_cost_by_cuisine
```

```python
avg_cost_by_cuisine.plot(kind='barh',title='Average Cost by Cuisine Type',ylabel='Cuisine Type',xlabel='Average Cost')
plt.show()
```

```python
#Proportion of Restaurants Offering Online Orders or Table Booking
#Group by area or restaurant type and calculate the proportion of Yes in online_order or table booking.
#Insight: How prevalent are online orders and table bookings in different areas?
```

```python
online_table_proportion=df.groupby('area')[['online_order','table booking']].apply(lambda x:(x=='Yes').mean())
```

```python
online_table_proportion
```

```python
online_table_proportion.plot(kind='barh',title='Proportion of Restaurants Offering Online Orders or Table Booking',ylabel='Area',xlabel='Proportion')
plt.show()
```

```python
#Analyze Binary Columns (online_order and table booking)
#Melt these columns into a long format with restaurant name or area to see trends in online ordering and table booking.
#Insight: Which areas or restaurant types are more tech-friendly?
```

```python
binary_melt=pd.melt(df,id_vars=['restaurant name','area'],value_vars=['online_order','table booking'])
binary_melt
```

```python
binary_trend=binary_melt.groupby(['area','variable','value']).size().unstack().fillna(0)[:20]
binary_trend
```

```python
binary_trend.plot(kind='bar',stacked=True,title='trend in online orders and table booknig by area')
plt.show()
```
